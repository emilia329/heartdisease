```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns', None)
df= pd.read_csv('Heart_Disease_Prediction.csv')
```

```python
df
```

```python
def classifier(df):
    if (df['Heart Disease'] == 'Presence'):
        return 1
    else: 
        return 2
   
```

```python
df['heartclass']= df.apply(classifier,axis=1)
```

```python
def cholesterol(df):

    if df['Cholesterol'] < 200:
        return("good")
    elif df['Cholesterol'] > 240:
        return("High")
    else:
        return("Borderline")
```

```python
df['cholesterolclass']= df.apply(cholesterol,axis=1)
```

```python
df['cholesterolclass'].value_counts()
```

```python
df
```

```python
def blood(df):

    if df['BP'] > 140:
        return("high")
    else:
        return("good")
```

```python
df['bloodpressure']= df.apply(blood,axis=1)
```

```python
df
```

```python
def senior(df):

    if df['Age'] > 65:
        return("senior")
    else:
        return("nonsenior")
```

```python
df['seniorcitizen']= df.apply(senior,axis=1)
```

```python
df
```

```python
df['seniorcitizen'].value_counts().nlargest(n=5).plot(kind='bar',color='moccasin',title='Installer')
```
